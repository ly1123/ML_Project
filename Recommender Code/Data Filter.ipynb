```python
y_label = []
for i in range(df_new.shape[0]):
    for target_label in target_list:
        if df_new.iloc[i,:][target_label] == 1:
            y_label.append(target_map[target_label])
            user_id = df_new.iloc[i,:]['ncodpers']
            df_new.loc[df.ncodpers == user_id, target_label] = 0
            break
```

```python
import xgboost as xgb
```

```python
df_new = df_new.drop(['fecha_dato', 'ncodpers', 'fecha_alta', 'ult_fec_cli_1t'], axis=1)
```

```python
for column_name in df_new.columns:
    df_new[column_name] = pd.to_numeric(df_new[column_name], errors='coerce')
```

```python
df_new.dtypes
```

```python
x_train = df_new.iloc[:, :20]
y_train = df_new.iloc[:, -1]
```

```python
dtrain = xgb.DMatrix(x_train, label=y_train)
```

```python
param = {}
param['objective'] = 'multi:softprob'
param['eta'] = 0.05
param['max_depth'] = 8
param['silent'] = 1
param['num_class'] = 24
param['eval_metric'] = 'mlogloss'
param['min_child_weight'] = 1
param['subsample'] = 0.7
param['colsample_bytree'] = 0.7
param['seed'] = 0
num_rounds = 30
```

```python
bst = xgb.train(param, dtrain, num_rounds)
```
