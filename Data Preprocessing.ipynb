```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
path = "Data/train_ver2.csv"
df = pd.read_csv(path)
```

```python
# Check the missing values
df.isnull().any()
```

Change the date format in fecha_dato to the standard data types

```python
# fecha_dato
df['fecha_dato'] = pd.to_datetime(df['fecha_dato'], format='%Y-%m-%d', errors='coerce')
df.fecha_dato.unique()
```

Convert the category datatypes to integers

```python
# ind_empleado
df.ind_empleado.unique()
df['ind_empleado'] = df['ind_empleado'].fillna(value=0)
df.loc[df.ind_empleado == 'A', 'ind_empleado'] = 1
df.loc[df.ind_empleado == 'B', 'ind_empleado'] = 2
df.loc[df.ind_empleado == 'F', 'ind_empleado'] = 3
df.loc[df.ind_empleado == 'N', 'ind_empleado'] = 4
df.loc[df.ind_empleado == 'S', 'ind_empleado'] = 5
df['ind_empleado'] = pd.to_numeric(df['ind_empleado'], errors='coerce')
df['ind_empleado'] = df['ind_empleado'].astype(int)
df.head()
```

```python
# pais_residencia
df.loc[df.pais_residencia.isnull(), 'pais_residencia'] = 'UNKNOWN'
country_list = df.pais_residencia.unique()
for i in range(len(country_list)):
    df.loc[df.pais_residencia == country_list[i], 'pais_residencia'] = i
df.head()
```

```python
# sexo
df.dtypes
df.sexo.unique()
df['sexo'] = df['sexo'].fillna(value=2)
df.loc[df.sexo == 'H', 'sexo'] = 0
df.loc[df.sexo == 'V', 'sexo'] = 1
df['sexo'] = pd.to_numeric(df['sexo'], errors='coerce')
df['sexo'] = df['sexo'].astype(int)
df.head()
```

```python
# age
df.dtypes
df['age'].unique()
df.loc[df.age == ' NA', 'age'] = 0
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['age'] = df['age'].astype(int)
df.head()
```

```python
# fecha_alta
df['fecha_alta'] = pd.to_datetime(df['fecha_alta'], format='%Y-%m-%d', errors='coerce')
df.head()
```

```python
# ind_nuevo
df.ind_nuevo.unique()
df['ind_nuevo'] = df['ind_nuevo'].fillna(value=1)
df['ind_nuevo'] = df['ind_nuevo'].astype(int)
df.head()
```

```python
# antiguedad
df.antiguedad.unique()
df.loc[df.antiguedad == '     NA', 'antiguedad'] = 0
df['antiguedad'] = pd.to_numeric(df['antiguedad'], errors='coerce')
df['antiguedad'] = df['antiguedad'].astype(int)
df.loc[df.antiguedad < 0, 'antiguedad'] = 0
df.head()
```

```python
# indrel
df.indrel.unique()
df['indrel'] = df['indrel'].fillna(value=0)
df['indrel'] = df['indrel'].astype(int)
df.head()
```

```python
# ult_fec_cli_1t
df.ult_fec_cli_1t.unique()
df['ult_fec_cli_1t'] = pd.to_datetime(df['ult_fec_cli_1t'], format='%Y-%m-%d', errors='coerce')
df.head()
```

```python
# indrel_1mes
df.indrel_1mes.unique()
df['indrel_1mes'] = df['indrel_1mes'].fillna(value=0)
df.loc[df.indrel_1mes == 'P', 'indrel_1mes'] = 5
df['indrel_1mes'] = pd.to_numeric(df['indrel_1mes'], errors='coerce')
df['indrel_1mes'] = df['indrel_1mes'].astype(int)
df.head()
```

```python
# tiprel_1mes
df.tiprel_1mes.unique()
df['tiprel_1mes'] = df['tiprel_1mes'].fillna(value=0)
df.loc[df.tiprel_1mes == 'A', 'tiprel_1mes'] = 1
df.loc[df.tiprel_1mes == 'I', 'tiprel_1mes'] = 2
df.loc[df.tiprel_1mes == 'P', 'tiprel_1mes'] = 3
df.loc[df.tiprel_1mes == 'R', 'tiprel_1mes'] = 4
df.loc[df.tiprel_1mes == 'N', 'tiprel_1mes'] = 5
df['tiprel_1mes'] = pd.to_numeric(df['tiprel_1mes'], errors='coerce')
df['tiprel_1mes'] = df['tiprel_1mes'].astype(int)
df.head()
```

```python
# indresi
df.indresi.unique()
df['indresi'] = df['indresi'].fillna(value=0)
df.loc[df.indresi == 'S', 'indresi'] = 1
df.loc[df.indresi == 'N', 'indresi'] = 2
df['indresi'] = pd.to_numeric(df['indresi'], errors='coerce')
df['indresi'] = df['indresi'].astype(int)
df.head()
```

```python
# indext
df.indext.unique()
df['indext'] = df['indext'].fillna(value=0)
df.loc[df.indext == 'S', 'indext'] = 1
df.loc[df.indext == 'N', 'indext'] = 2
df['indext'] = pd.to_numeric(df['indext'], errors='coerce')
df['indext'] = df['indext'].astype(int)
df.head()
```

```python
# conyuemp
df.conyuemp.unique()
df['conyuemp'] = df['conyuemp'].fillna(value=0)
df.loc[df.conyuemp == 'S', 'conyuemp'] = 1
df.loc[df.conyuemp == 'N', 'conyuemp'] = 2
df['conyuemp'] = pd.to_numeric(df['conyuemp'], errors='coerce')
df['conyuemp'] = df['conyuemp'].astype(int)
df.head()
```

```python
# canal_entrada
df['canal_entrada'] = df['canal_entrada'].fillna(value='UNKNOWN')
temp_list = list(df.canal_entrada.unique())
for i in range(len(temp_list)):
    df.loc[df.canal_entrada == temp_list[i], 'canal_entrada'] = i
```

```python
# indfall
df.indfall.unique()
df['indfall'] = df['indfall'].fillna(value=0)
df.loc[df.indfall == 'S', 'indfall'] = 1
df.loc[df.indfall == 'N', 'indfall'] = 2
df['indfall'] = pd.to_numeric(df['indfall'], errors='coerce')
df['indfall'] = df['indfall'].astype(int)
df.head()
```

```python
# tipodom
df.tipodom.unique()
df['tipodom'] = df['tipodom'].fillna(value=0)
df['tipodom'] = df['tipodom'].astype(int)
df.head()
```

```python
# cod_prov
df.cod_prov.unique()
df['cod_prov'] = df['cod_prov'].fillna(value=0)
df['cod_prov'] = df['cod_prov'].astype(int)
df.head()
```

Since there are many missing values in the income, we break it down for the province and substitute the missing value with the mean of the corresponding province.

```python
# renta
df.loc[df.nomprov.isnull(), 'nomprov'] = 'UNKNOWN'
province_name = list(df.nomprov.unique())
for i in range(len(df.nomprov.unique())):
    mean = df.loc[df.nomprov == province_name[i]]['renta'].mean(skipna=True)
    df.loc[df.nomprov == province_name[i], 'renta'] = df.loc[df.nomprov == province_name[i], 'renta'].fillna(value=mean)
```

```python
# nomprov
df.nomprov.unique()
temp_list = list(df.nomprov.unique())
for i in range(len(temp_list)):
    df.loc[df.nomprov == temp_list[i], 'nomprov'] = i
df.head()
```

```python
# ind_actividad_cliente
df.ind_actividad_cliente.unique()
df['ind_actividad_cliente'] = df['ind_actividad_cliente'].fillna(value=2)
df['ind_actividad_cliente'] = df['ind_actividad_cliente'].astype(int)
df.head()
```

```python
# segmento
df.segmento.unique()
df['segmento'] = df['segmento'].fillna(value=0)
df.loc[df.segmento == '01 - TOP', 'segmento'] = 1
df.loc[df.segmento == '02 - PARTICULARES', 'segmento'] = 2
df.loc[df.segmento == '03 - UNIVERSITARIO', 'segmento'] = 3
df['segmento'] = pd.to_numeric(df['segmento'], errors='coerce')
df['segmento'] = df['segmento'].astype(int)
df.head()
```

```python
# ind_nomina_ult1
df.ind_nomina_ult1.unique()
df['ind_nomina_ult1'] = df['ind_nomina_ult1'].fillna(value=0)
df['ind_nomina_ult1'] = df['ind_nomina_ult1'].astype(int)
df.head()
```

```python
# ind_nom_pens_ult1
df.ind_nom_pens_ult1.unique()
df['ind_nom_pens_ult1'] = df['ind_nom_pens_ult1'].fillna(value=0)
df['ind_nom_pens_ult1'] = df['ind_nom_pens_ult1'].astype(int)
df.head()
```
